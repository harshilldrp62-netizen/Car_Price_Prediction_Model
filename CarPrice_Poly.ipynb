```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv('cleaned_carprice.csv')
```

```python
df['year'].astype('int')
```

```python
x = df.drop(columns='Price')
y = df['Price']
```

```python
poly = PolynomialFeatures(degree=5)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=989)
```

```python
ohe = OneHotEncoder(handle_unknown='ignore')
ohe.fit(x[['name','company','fuel_type']])
```

```python
column_trans1 = make_column_transformer((OneHotEncoder(categories=ohe.categories_,sparse_output=False),['name','company','fuel_type']),remainder = 'passthrough',verbose_feature_names_out=False)
column_trans1.set_output(transform="pandas")
column_trans2 = make_column_transformer((PolynomialFeatures(degree=2),['year','kms_driven']),remainder='passthrough')
column_trans1.set_output(transform="pandas")
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans1,column_trans2,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```
